```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
## simple example
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
## ploting using matplotlib
plt.scatter(x,y,c='g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
## compute the x and y  coordinates for  ponts on sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sine wave from')

# plot yhe points using matplotlib
plt.plot(x,y,'r')
plt.show()
```

```python
# subplot()
# compute the x and y coordinate for points on sin and cosine curves
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#Set up a subplot grid that has height 2 and width 1,
#and set the 1st  such subplot  a sactive
plt.subplot(2,1,1)
# make the 1st plot
plt.plot(x,y_sin)
plt.title('sine')

# Set the scoend subplot as active ,and make the scenod plot.
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cosine')
# show the figure
plt.show()
```

```python
## Bar plot
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Barplot')
plt.xlabel('x axies')
plt.ylabel('y axies')
plt.show()
```

```python
## Histogram
a=np.array([22,34,67,90,24,67,84,78,57,76,70,94])
plt.hist(a)
plt.title('histogram')
plt.show()
```

```python
# box plot using matplot
data=[np.random.normal(0, std ,100) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
# pie chart
# data to plot
labels='python','c++','Ruby','java'
sizes = [215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) # explod 1st slice

#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
